**Getting required libraries**

```python
import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
import re

# Download required resources
nltk.download('vader_lexicon')
nltk.download('punkt')
nltk.download('stopwords')

# Load your review data into Python
positive_sentences_array = positive_strings_array  # All the sentences as a string array ----- INPUT -----
negative_sentences_array = negative_strings_array  # All the sentences as a string array ----- INPUT -----
```

**Prepeocessing the each review**

```python
# Clean the text by removing unnecessary characters, numbers, or special symbols
def clean_text(text):
    text = re.sub(r'\W', ' ', text)  # Remove non-word characters
    text = re.sub(r'\s+', ' ', text)  # Remove extra whitespaces
    text = text.lower().strip()  # Convert to lowercase and remove leading/trailing whitespaces
    return text

# Tokenize the reviews into sentences using NLTK's sent_tokenize function
def tokenize_sentences(review):
    sentences = sent_tokenize(review)
    return sentences

# Remove stop words from sentences
def remove_stopwords(sentence):
    stop_words = set(stopwords.words('english'))
    sentence = ' '.join([word for word in sentence.split() if word.lower() not in stop_words])
    return sentence

# Apply data preprocessing steps to the sentences_array
def Return_preprocessed_sentences(sentences_array):
  preprocessed_sentence_array = []
  for review in sentences_array:
      review = clean_text(review)
      sentences = tokenize_sentences(review)
      sentences = [remove_stopwords(sentence) for sentence in sentences]
      preprocessed_sentence_array.append(sentences)
  return preprocessed_sentence_array
```

**This function analyses the Sentiment and calculate the Positivity and Negativity.**

```python
# Function to analyze the sentiment of a sentence
def analyze_sentiment(sentence):
    sid = SentimentIntensityAnalyzer()
    sentiment_scores = sid.polarity_scores(sentence)
    sentiment = 'positive' if sentiment_scores['compound'] >= 0 else 'negative'
    positivity = sentiment_scores['pos']
    negativity = sentiment_scores['neg']
    return sentiment, positivity, negativity
```

**This function takes an array of strings and create 3 arrays (sentiment, positivity and negativity) using each string inside array and returns.**

```python
# Extract sentiments, positivity percentages, and negativity percentages as arrays
def Return_Sentiments(preprocessed_sentences):
  sentiments = []
  positivity_percentages = []
  negativity_percentages = []
  for review_sentences in preprocessed_sentences:
      if len(review_sentences) == 0:
          sentiments.append('positive')
          positivity_percentages.append(0)
          negativity_percentages.append(0)
      else:
          positivity_sum = 0
          negativity_sum = 0
          for sentence in review_sentences:
              sentiment, positivity, negativity = analyze_sentiment(sentence)
              positivity_sum += positivity
              negativity_sum += negativity

          total_sentences = len(review_sentences)
          positivity_percentage = (positivity_sum / total_sentences) * 100
          negativity_percentage = (negativity_sum / total_sentences) * 100

          sentiments.append(sentiment)
          positivity_percentages.append(positivity_percentage)
          negativity_percentages.append(negativity_percentage)

  return {
      "sentiment":   sentiments,
      "positivity_percentages": positivity_percentages,
      "negativity_percentages": negativity_percentages
  }
```

**Returned Dictionaries are stored here**

```python
# Creating the preprocessed sentence arrays
positive_preprocessed_sentences = Return_preprocessed_sentences(positive_sentences_array)
negative_preprocessed_sentences = Return_preprocessed_sentences(negative_sentences_array)

print(positive_preprocessed_sentences)
print(negative_preprocessed_sentences)
```

**Create a Dataframe with sentiment data**

```python
positive_sentence_column_sentiments = Return_Sentiments(positive_preprocessed_sentences)
negative_sentence_column_sentiments = Return_Sentiments(negative_preprocessed_sentences)

# Create a DataFrame with the preprocessed data
sentiments_df = pd.DataFrame({
    'positive_sentence_column_Review': positive_sentences_array,
    'positive_sentence_column_Sentiment': positive_sentence_column_sentiments["sentiment"],
    'positive_sentence_column_Positivity Percentage': positive_sentence_column_sentiments["positivity_percentages"],
    'positive_sentence_column_Negativity Percentage': positive_sentence_column_sentiments["negativity_percentages"],
    'negative_sentence_column_Review': negative_sentences_array,
    'negative_sentence_column_Sentiment': negative_sentence_column_sentiments["sentiment"],
    'negative_sentence_column_Positivity Percentage': negative_sentence_column_sentiments["positivity_percentages"],
    'negative_sentence_column_Negativity Percentage': negative_sentence_column_sentiments["negativity_percentages"],

    })

# Display the DataFrame
display(sentiments_df)
```
